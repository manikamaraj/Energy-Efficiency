```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dataset=pd.read_csv("ENB.csv")
X=dataset.iloc[:,:-2].values
y=dataset.iloc[:,8:10].values
```

```python
print(X)
```

```python
print(y)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import PolynomialFeatures
PF_regressor=PolynomialFeatures(degree=2)
X_poly=PF_regressor.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_poly,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
poly_regressor=LinearRegression()
poly_regressor.fit(X_train,y_train)
y_pred=poly_regressor.predict(X_test)
```

```python
np.set_printoptions(precision=2)
```

```python
result=np.concatenate((y_test,poly_regressor.predict(X_test)),axis=1)
```

```python
print(result)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
